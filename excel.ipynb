```python
import numpy as np
import pandas as pd
```

```python
legajo_tarjeta = pd.read_excel('legajo_reloj.xls', )
legajo_tarjeta
```

```python
usuario_reloj = pd.read_excel('usuario_datos.xlsx', )
usuario_reloj
```

```python
inner = pd.merge(left=legajo_tarjeta, right=usuario_reloj, left_on='Nro Identificador', right_on='legajo')
inner
```

## Extract useful data

```python
merged = inner[['Nro Identificador', 'Nro Tarjeta', 'apellido',
                    'nombre', 'fecnac', 'antiguedad', 'domicilio', 'nro']]
merged['nro'][merged['nro'].isnull()] = 0
merged['nro'][merged['nro'] == 0] 
merged['nro'] = merged['nro'].astype(np.intc)
```

## Concatenate last name and name, then delete the columns

```python

merged['Nombre'] = merged.apply(
    lambda x: x['apellido'] + ' ' + x['nombre'], axis=1)
del merged['nombre']
del merged['apellido']
```

## Concatenate address and number, the delete the columns

```python
merged['Domicilio'] = merged.apply(
    lambda x: x['domicilio'] + ' ' + str(x['nro']), axis=1)
del merged['domicilio']
del merged['nro']
```

## Add user number

```python
length = merged.shape[0]
merged['Usuario Nro.'] = np.arange(1, length + 1)
```

## Add needed columns

```python
merged['GÃ©nero'] = np.nan
merged['Departamento'] = 'Hospital'
merged['PosiciÃ³n'] = np.nan
merged['TelÃ©fono'] = np.nan
merged['Dispositivos'] = np.nan
```

## Correct columns types

```python
merged['fecnac'] = merged['fecnac'].astype(str)
for el in merged['fecnac']:
    el = el[6:16]
merged['antiguedad'] = merged['antiguedad'].astype(str)
for el in merged['antiguedad']:
    el = el[6:16]
```

## Rename columns

```python
merged = merged.rename(columns={"Nro Identificador": "ID en dispositivo", "Nro Tarjeta": "Tarjeta Nro.", "fecnac": "CumpleaÃ±os", "antiguedad": "Fecha de inicio laboral", "B": "c"})
```

## Generate final output

```python
final_dataframe = merged[['Usuario Nro.', 'ID en dispositivo', 'Tarjeta Nro.', 'Nombre',
                              'Departamento', 'GÃ©nero', 'PosiciÃ³n', 'CumpleaÃ±os', 'Fecha de inicio laboral', 'TelÃ©fono', 'Domicilio', 'Dispositivos']]
final_dataframe.fillna('', inplace=True)
final_dataframe.loc[final_dataframe['CumpleaÃ±os'] == 'NaT', 'CumpleaÃ±os'] = '1990-01-01'
final_dataframe.to_excel('Reloj Hospital.xlsx', index=False)
```
