```python
import numpy as np
import pandas as pd
```

```python
legajo_tarjeta = pd.read_excel('ListadoLegajos20-1-2023 8.37.24.xls', )
legajo_tarjeta
```

```python
usuario_reloj = pd.read_excel('usuario_reloj.xlsx', )
usuario_reloj
```

```python
inner = pd.merge(left=legajo_tarjeta, right=usuario_reloj, left_on='Nro Identificador', right_on='legajo')
inner
```

## Extract useful data

```python
for_export = inner[['Nro Identificador', 'Nro Tarjeta', 'apellido',
                    'nombre', 'fecnac', 'antiguedad', 'domicilio', 'nro']]
for_export['nro'][for_export['nro'].isnull()] = 0
for_export['nro'][for_export['nro'] == 0] 
for_export['nro'] = for_export['nro'].astype(np.intc)
```

## Concatenate last name and name, then delete the columns

```python

for_export['Nombre'] = for_export.apply(
    lambda x: x['apellido'] + ' ' + x['nombre'], axis=1)
del for_export['nombre']
del for_export['apellido']
```

## Concatenate address and number, the delete the columns

```python
for_export['Domicilio'] = for_export.apply(
    lambda x: x['domicilio'] + ' ' + str(x['nro']), axis=1)
del for_export['domicilio']
del for_export['nro']
```

## Add user number

```python
length = for_export.shape[0]
for_export['Usuario Nro.'] = np.arange(1, length + 1)
```

## Add needed columns

```python
for_export['Género'] = np.nan
for_export['Departamento'] = 'Hospital'
for_export['Posición'] = np.nan
for_export['Teléfono'] = np.nan
for_export['Dispositivos'] = np.nan
```

## Correct columns types

```python
for_export['fecnac'] = for_export['fecnac'].astype(str)
for el in for_export['fecnac']:
    el = el[6:16]
for_export['antiguedad'] = for_export['antiguedad'].astype(str)
for el in for_export['antiguedad']:
    el = el[6:16]
```

## Rename columns

```python
for_export = for_export.rename(columns={"Nro Identificador": "ID en dispositivo", "Nro Tarjeta": "Tarjeta Nro.", "fecnac": "Cumpleaños", "antiguedad": "Fecha de inicio laboral", "B": "c"})
```

## Generate final output

```python
final_dataframe = for_export[['Usuario Nro.', 'ID en dispositivo', 'Tarjeta Nro.', 'Nombre',
                              'Departamento', 'Género', 'Posición', 'Cumpleaños', 'Fecha de inicio laboral', 'Teléfono', 'Domicilio', 'Dispositivos']]
final_dataframe.fillna('', inplace=True)
final_dataframe.loc[final_dataframe['Cumpleaños'] == 'NaT', 'Cumpleaños'] = '1990-01-01'
final_dataframe.to_excel('Reloj Hospital.xlsx', index=False)
final_dataframe[17:45][['Cumpleaños', 'Nombre']]
```
